```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
arr = iris.values
```

```python
arr.shape
```

```python
iris_data = arr[:150,1:5]       #features
iris_type = arr[:150,5]         #labels
```

```python

for i in range(iris_type.size):
            if iris_type[i]=='Iris-setosa':
                   iris_type[i]=0
            elif iris_type[i]=='Iris-versicolor':
                    iris_type[i]=1
            elif iris_type[i]=='Iris-virginica':
                iris_type[i]=2
                    
        
        
        # in here we do string to int convert because in scikit learn model fit as labels in float values
```

```python
iris_type
```

```python
reg = linear_model.LinearRegression()      # create linear regression object
```

```python
model = reg.fit(iris_data,iris_type)                #fit training model using training sets
```

```python
plt.plot(iris_data,iris_type)
```

```python
model.score(iris_data,iris_type)   # it return accuracy of model
```

```python
predict = model.predict(np.array([[5.9, 3.0, 5.1, 1.8]]))
predict
```
